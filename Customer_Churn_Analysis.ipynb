## Churn Distribution

This chart shows the number of customers who stayed and those who churned.

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Churn", data=df)

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")

plt.show()
```

### Insight

The chart shows the distribution of customers who stayed and those who churned. This provides an overall understanding of customer retention in the dataset.

## Contract Length vs Churn

```python
plt.figure(figsize=(8,5))

sns.countplot(x="Contract Length", hue="Churn", data=df)

plt.title("Contract Length vs Churn")
plt.xlabel("Contract Length")
plt.ylabel("Customer Count")

plt.show()
```

### Insight

This visualization helps understand whether customers with different contract lengths are more likely to churn.

## Subscription Type vs Churn

```python
plt.figure(figsize=(8,5))

sns.countplot(x="Subscription Type", hue="Churn", data=df)

plt.title("Subscription Type vs Churn")
plt.xlabel("Subscription Type")
plt.ylabel("Customer Count")

plt.show()
```

### Insight

The chart compares churn across different subscription plans and identifies which subscription type experiences higher customer churn.

## Total Spend vs Churn

This boxplot compares the spending behavior of customers who churned and those who stayed.

```python
plt.figure(figsize=(7,5))

sns.boxplot(x="Churn", y="Total Spend", data=df)

plt.title("Total Spend vs Churn")
plt.xlabel("Churn")
plt.ylabel("Total Spend")

plt.tight_layout()
plt.show()
```

### Insight

This chart compares the total spending of churned and retained customers. It helps understand whether spending behavior is associated with customer churn.

## Correlation Heatmap

The heatmap shows the relationship between numerical variables in the dataset.

```python
plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include=["int64","float64"])

sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")

plt.title("Correlation Heatmap")

plt.tight_layout()
plt.show()
```

### Insight

The heatmap displays the correlation between numerical features. It helps identify which variables have stronger positive or negative relationships with customer churn.
